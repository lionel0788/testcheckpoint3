```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

```python
plot_correlation_map( data)
```

```python
# analyse de la fonction et ses utilit√©s:
#l'on remarque qu'il y a 100% de corr√©lation entre les variables qui se ressemblent.
#il y a aussi une corr√©lation de 41% entre le Parch et le SibSp,les autres corr√©lations sont faibles(-25% de corr√©lation)
#Traduction en anglais
#These data represent the different correlations between the data in the dataset.
#These correlations are explained by percentages.
#The colors therefore identify the level of correlation based on the percentage.
```
